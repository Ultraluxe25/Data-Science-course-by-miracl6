```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

# Заголовок
ax[0].set_title(label='Синие точки', fontsize=20)
ax[1].set_title(label='Красные точки', fontsize=20)

# Точки на графике
ax[0].scatter(x=[0.6, 1.5], y=[0.0, -0.5], color='darkblue')
ax[1].scatter(x=[1.2, 0.7], y=[-0.1, 0.1], color='red')

# Подписи по осям
ax[0].set_xlabel(xlabel='OX', fontsize=16)
ax[0].set_ylabel(ylabel='OY', fontsize=16)
ax[1].set_xlabel(xlabel='OX', fontsize=16)
ax[1].set_ylabel(ylabel='OY', fontsize=16)

# Настройка параметров компановки
plt.subplots_adjust(wspace=0.3)
plt.show()
```

В plt.subplots() мы указали количество строк и столбцов соответственно, на которые хотим разбить наш рисунок Figure

Чтобы сделать заголовки/подписи/настройку делений и так далее для каждого из графиков, необходимо получить доступ к каждой из области рисования Axes. plt.subplots() возвращает рисунок Figure и массив из областей Axes, поэтому получив доступ к конкретной области по ее индексу, мы можем обратиться к методам, которые позволяют сделать тонкую настройку. Как правило с приставкой set (например set_title)

Чтобы настроить расстояние между графиками (по ширине/длине и так далее) используют метод plt.subplots_adjust(). При указании значений для определенных параметров, вы можете настраивать по вашему усмотрению рисунок. В нашем случае мы увеличили расстояние между графиками, задав wspace=0.3

wspace - ширина отступа между подзаголовками, как доля от средней ширины осей

**Scatterplot** (диаграмма рассеяния) может выявить симметрию данных, кластеры, корреляцию между переменными, а также экстремальные значения или выбросы.

# [seaborn.scatterplot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html)

**sns.heatmap()** - "тепловая карта", может применяться в анализе корреляций как положительных, так и отрицательных, анализе значений значимых признаков, которые могут повлиять на целевую переменную при использовании модели ML.

```python
sns.heatmap(df.corr(), annot=True, cmap='crest');
```

При помощи **sns.heatmap()** удобнее именно оценивать численно сразу несколько признаков и целевую переменную между собой.
