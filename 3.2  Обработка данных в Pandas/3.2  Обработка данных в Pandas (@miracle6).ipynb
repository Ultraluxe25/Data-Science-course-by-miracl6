````python
orders = pd.DataFrame({
    'id': [0, 1, 2, 3, 4],
    'id_product': [1, 0, 2, 0, 0],
    'count': [2, 10, 5, 1, 1]
})

products = pd.DataFrame({
    'id_product': [0, 1, 2],
    'price': [100, 75, 235]
})
````

````python
orders
````

````python
products
````

````python
df_sum = pd.merge(orders, products, on='id_product', how='left')
df_sum
````

````python
df_sum['revenue'] = df_sum['count'] * df_sum['price']
df_sum
````
